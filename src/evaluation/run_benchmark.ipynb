# Run the BenchmarkRunner in a Notebook
This notebook demonstrates how to run the `BenchmarkRunner` class to evaluate models on a dataset.

```python
# Add the evaluation directory to the Python path
import sys
from pathlib import Path

evaluation_dir = Path("/workspace/src/evaluation")
if str(evaluation_dir) not in sys.path:
    sys.path.append(str(evaluation_dir))
```

```python
# Import the BenchmarkRunner class
from benchmark_runner import BenchmarkRunner
import asyncio
```

```python
# Define input parameters
qa_pairs_path = "/workspace/data/qa_pairs.csv"  # Path to the QA pairs CSV file
output_dir = "/workspace/results"              # Directory to save results
models_to_test = ["gpt-4o", "llama-3.3-70b"]   # List of models to evaluate
strategy = "model_by_model"                    # Execution strategy: 'model_by_model' or 'row_by_row'
batch_size = 25                                # Number of rows to process before saving progress
max_concurrency = {"openai": 5}                # Max concurrency per provider (optional)
```

```python
# Initialize the BenchmarkRunner
runner = BenchmarkRunner(
    models_to_test=models_to_test,
    strategy=strategy,
    batch_size=batch_size,
    max_concurrency=max_concurrency,
)
```

```python
# Run the benchmark using the existing event loop
async def run_benchmark():
    await runner.run(qa_pairs_path, output_dir)

# Use asyncio's event loop to run the coroutine
await run_benchmark()
```

# Check Results
The results will be saved in the specified `output_dir` as separate CSV files for each model.

```python
# List the generated result files
import os

result_files = os.listdir(output_dir)
print("Generated result files:", result_files)
```
